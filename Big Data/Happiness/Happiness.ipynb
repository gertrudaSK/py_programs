```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
feature_cols = ['Life Expectancy', 'Marriage rate', 'Divorce rate', 'Unemployment rate',
       'Fossil fuels', 'Diabetes Prevalance', 'Rights Strength',
       'Women in Parliament', 'Car Accidents', 'Urban Population']
```

```python
X = data[feature_cols]
y = data['Score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train, y_train)
```

```python
pred = linear.predict(X_test)
pred
```

```python
predictions = pd.Series(data=pred, name="Prediction")
```

```python
res = pd.concat([y_test.reset_index(), predictions], axis=1)[['Score', 'Prediction']]
```

```python
res.head() #spejimai ir tikrosios reiksmes
```

```python
linear.score(X_test, y_test) #modelio tikslumas
```

```python
sns.distplot(y_test, bins = 15) #patikrinam ar modelis pavyko ir duomenu issidestymas normalus
```

```python
coef = pd.DataFrame(data = linear.coef_, columns = ['Coefficient'], index = X_train.columns).sort_values(by=['Coefficient'],  ascending=False)
```
